+ Rename collumns name

```python
new_cols = {
    "timestamp": "time",
    "cnt": "count",
    "t1": "temp_real_C",
    "t2": "temp_feels_like_C",
    "hum": "humidity_percent",
    "wind_speed": "wind_speed_kph",
    "weather_code": "weather",
    "is_holiday": "is_holiday",
    "is_weekend": "is_weekend",
    "season": "season"
}
```

```python
df.rename(new_cols,axis=1,inplace=True)
```

```python
df
```

```python
df.humidity_percent=df.humidity_percent/100
```

```python
df
```

```python
# Creating a season dictionary so that
season_dict = {
    "0.0": "spring",
    "1.0": "summer",
    "2.0": "autumn",
    "3.0": "winter"
}

# Creating a weather dictionary so that
weather_dict = {
    "1.0": "Clear",
    "2.0": "Scattered clouds",
    "3.0": "Broken clouds",
    "4.0": "Cloudy",
    "7.0": "Rain",
    "10.0": "Rain with thunderstorm",
    "26.0": "Snowfall"
}
```

```python
df.season=df.season.astype('str')
df.season=df.season.map(season_dict)
```

```python
df.weather=df.weather.astype('str')
df.weather=df.weather.map(weather_dict)
```
